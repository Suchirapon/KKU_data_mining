#Box plot

```
from matplotlib import pyplot as plt
```

```
plt.boxplot(data['x'])
```


ใช้คำสั่ง __vert__ เพื่อในกล่องอยู่ในแนวนอน

```
plt.boxplot(data['x'],vert=False)
```

เปลี่ยน whisker เป็น 1.2 ให้มี ค่า out lier ให้มีมากกว่าเดิม

```
plt.boxplot(data['x'],vert=False,whis=1.2)
```

```
oput = plt.boxplot(data['x'],vert=False,whis=1.2)
```

ให้แสดงค่า outlier ถ้าอยากได้ค่าของอะไรให้ใส่ใน __[ ]__ ตามด้วย .get_xdata() โดยที่จะเอาค่า x หรือ y ก็ได้ ที่แสดงออกมาคือค่าที่เกิน

```
oput['fliers'][0].get_xdata()
```

เส้น whisker วาดจากค่า Q1(=44.1026) ถึง 1.2IQR(=22.3077) (1.2 เรากำหนดใหม่ ปกติเป็น 1.5)

```
oput['whiskers'][0].get_xdata()
```

```
oput['whiskers'][1].get_xdata()
```

# scatter plot

```
plt.scatter(data['x'],data['y'])
```

__s__ คือ การเพิ่มค่าหรือลดค่า ขนาดของแต่ละจุด

```
plt.scatter(data['x'],data['y'],s=25)
```

__s=data['y']__ ให้จุดเปลี่ยนขนาดตามค่า y

```
plt.scatter(data['x'],data['y'],s=data['y'])
```

__c__ คือ การทำให้อาเรย์ หรือ ลิตส์ มีค่าสีของ RGB,CMYK

```
plt.scatter(data['x'],data['y'],s=data['y'],c='g')
```

__marker__ คือ การเปลี่ยนจุดให้เป็นรูปแบบต่างๆ

```
plt.scatter(data['x'],data['y'],s=data['y'],c='m',marker='1')
```

__alpha__ คือ ความโปร่งแสงของเลเยอร์ เริ่ม 0 คือ โปร่งแสง จนถึง 1 คือ ทึบแสง

```
plt.scatter(data['x'],data['y'],s=data['y'],c='r',marker='1',alpha=0.4)
plt.scatter([44,42,50],[15,14,80],s=250,alpha=0.6)
```
